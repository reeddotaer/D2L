```python
import numpy as np
def sample(n):
    instance = np.random.randint(2, size=(n,))
    parity = instance.sum()%2
    return instance, parity

def instance_to_tensor(instance):
    tensor = torch.zeros(len(instance),1,2)
    for i in range(len(instance)):
        bit = instance[i]
        tensor[i][0][bit] = 1
    return tensor

def label_to_tensor(label):
    tensor = torch.zeros(1,2)
    tensor[0][label] = 1
    return tensor


def fetch_x_y(n):
    instance,label = sample(n)
    x = instance_to_tensor(instance)
    y = label_to_tensor(label)
    return x,y
```

```python
import torch.nn as nn
# import pytorch as torch
import torch


class RNN(nn.Module):
 
    def __init__(self, input_size, hidden_size, output_size, num_layer):
        super(RNN, self).__init__()
        
        self.hidden_size = hidden_size
        self.num_layer = num_layer
                
        self.i2h = nn.Linear(input_size + hidden_size, hidden_size)
        self.i2o = nn.Linear(input_size + hidden_size, output_size)
        self.softmax = nn.LogSoftmax(dim = 1)
        
    def forward(self, input):
        hidden = self.hidden
        for i in range(self.num_layer):
            combined = torch.cat((input[i], hidden), 1)
            hidden = self.i2h(combined)
#             hidden = self.softmax(hidden)
        output = self.i2o(combined)
        output = self.softmax(output)
        return output
    
    def init_hidden(self):
        init = torch.zeros(1,self.hidden_size)
        init[0][0] = 1
        self.hidden = init
        return 
    

n_hidden = 4
n_length = 3
n_catagories = 2

rnn = RNN(2, n_hidden, n_catagories,n_length)

# for param in rnn.parameters():
#     print(param)
# print(rnn.parameters())
# print(2, n_hidden, n_catagories, n_length)
```

```python
loss_func = nn.CrossEntropyLoss()
learning_rate = 0.005 
optimizer = torch.optim.SGD(rnn.parameters(), lr=learning_rate)

def train(x,y):
    rnn.zero_grad()
    rnn.init_hidden()
    output = rnn.forward(x)

    loss = loss_func(y,output)
    loss.backward()

    optimizer.step()
    return output, loss
```
